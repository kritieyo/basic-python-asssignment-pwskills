```python
'''
8. What is a hash table, and how does it relate to dictionaries in Python?

A hash table is a data structure that stores data using key–value pairs and uses a hash function to quickly find where each value is stored.

In Python, a dictionary is implemented using a hash table.
This is why dictionary operations like lookup, insert, and delete are very fast — they use hashing to jump directly to where the key’s value is stored.
```

```python
'''
10. Explain why strings are immutable in Python.

Strings are immutable in Python because they cannot be changed once created. This immutability makes Python more efficient, since unchanged string objects can be safely reused in memory, improving performance. It also provides safety by preventing accidental modifications to data, and it allows strings to be used as dictionary keys because only immutable objects can be reliably hashed.
```

```python
'''
17. Describe the time complexity of accessing elements in a dictionary.

Accessing elements in a dictionary has an average time complexity of O(1), meaning values can be retrieved instantly using their keys. This is because dictionaries use hashing, which allows fast lookups. In rare cases of hash collisions, the time complexity can degrade, but on average dictionaries are highly efficient.
```

```python
'''
18. In what situations are lists preferred over dictionaries?

Lists are preferred over dictionaries when the order of elements matters, when you need to access items by position or index, or when storing a simple sequence of values without key–value pairing. They are also better for tasks involving iteration in a specific order or when duplicates are allowed.
```

```python
'''
19. Why are dictionaries considered unordered, and how does that affect data retrieval?

Dictionaries are considered unordered because their elements are stored based on hash values, not on their insertion order (conceptually). As a result, you cannot rely on a specific positional order when storing data. However, this does not affect data retrieval, because values are accessed directly using their keys, which makes lookups fast and efficient regardless of order.
```
