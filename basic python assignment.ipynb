```python
'''
Basic python answers
THEORETICAL QUESTIONS (PYTHON)

1. What is Python, and why is it popular?

ANSWER:
Python is a high-level, interpreted programming language known for its readability and simplicity. It's popular because:
It has a simple syntax similar to English.
It's versatile. Used in web development, data science, automation, etc.
It has a large standard library and a supportive community


2. What is an interpreter in Python?

ANSWER:
An interpreter is a program that reads and executes Python code line by line and it translates Python code into machine code at runtime.


3. What are pre-defined keywords in Python?

ANSWER:
Pre-defined keywords are reserved words that have special meaning in Python, such as if, else, def, class, etc. They cannot be used as variable names.


4. Can keywords be used as variable names?

ANSWER:
No, Python keywords cannot be used as variable names. Doing so will result in a syntax error.


5. What is mutability in Python?

ANSWER:
Mutability refers to whether an object can be changed after it's created. Mutable objects (e.g., lists, dictionaries) can be modified.


6. Why are lists mutable, but tuples are immutable?

ANSWER:
Lists are designed for dynamic data and allow modifications (add, remove, update). Tuples are used for fixed data and do not allow changes to ensure data integrity and performance optimization.


7. What is the difference between “==” and “is” operators in Python?

ANSWER:
"==" checks value equality (whether two variables have the same value). While "is" checks identity (whether two variables point to the same memory location).



8. What are logical operators in Python?

ANSWER:
Logical operators combine multiple conditions:
and: True if both conditions are true,
or: True if at least one condition is true,
not: Reverses the logical state.


9. What is type casting in Python?

ANSWER:
Type casting is converting one data type into another, e.g., int("5") to convert a string to an integer.


10. What is the difference between implicit and explicit type casting?

ANSWER:
Implicit: Done automatically by Python (e.g., int + float → result is float)
Explicit: Manually done by the programmer using casting functions (e.g., float(5)).


11. What is the purpose of conditional statements in Python?

ANSWER:
Conditional statements like if, elif, and else help execute specific blocks of code based on certain conditions.


12. How does the elif statement work?

ANSWER:
elif (else if) allows checking multiple conditions after an if. If the if is False, Python evaluates elif blocks in order until one is True.


13. What is the difference between for and while loops?

ANSWER:
for loop: Iterates over a sequence (e.g., list, range)
while loop: Continues as long as a condition is True.


14. Describe a scenario where a while loop is more suitable than a for loop.

ANSWER:
Use a while loop when the number of iterations is unknown, such as reading user input until they enter "exit".
'''
```

```python
#5.Mutable vs Immutable arguments

def mutate_list(lst):
    lst.append(4)

def mutate_int(x):
    x += 1
    return x

my_list = [1, 2, 3]
my_num = 10

mutate_list(my_list)
print("List after function call:", my_list)  # Mutable

new_num = mutate_int(my_num)
print("Original int:", my_num)  # Immutable
print("Returned int:", new_num)
```

```python
#8. List vs Tuple mutation

my_list = [1, 2, 3]
my_tuple = (1, 2, 3)

my_list[0] = 10
print("Modified list:", my_list)

# This will raise an error
try:
    my_tuple[0] = 10
except TypeError as e:
    print("Error modifying tuple:", e)
```

```python
#14. Reverse a string using a while loop

text = input("Enter a string: ")
reversed_text = ""
i = len(text) - 1
while i >= 0:
    reversed_text += text[i]
    i -= 1
print("Reversed string:", reversed_text)
```

```python
#15. Factorial using a while loop

num = int(input("Enter a number: "))
factorial = 1
i = 1
while i <= num:
    factorial *= i
    i += 1
print(f"Factorial of {num} is {factorial}")
```
